#### Here also we have colinearity, that verticle line in the above graph is due to the fact that nan value in people hospitalized have been replaced by its mean.

```python
y1=df_final['Deaths']
y2=df_final['Recovered']
x=df_final.drop(['Deaths','Recovered'],axis=1)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
ss=StandardScaler()
```

```python
x=ss.fit_transform(x)
```

```python
y1=scale(y1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
from sklearn.neural_network import MLPRegressor
nn=MLPRegressor(hidden_layer_sizes=(49,7),max_iter=1000)
lr=LinearRegression()
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR
svr=SVR()
```

#### In the neural network we assumed that the independent variables to have maximum 8 principal components(out of 9 columns) and hence have 7 free variables for each instances. 7 free variables can make maximally 49 first hidden layer neuron and for convergence we took 7 neurons in 2nd hidden layer.  

```python
X_train,X_test,y_train,y_test=train_test_split(x,y1,test_size=0.20)
models=[lr,tree,knr,nn,svr]
for model in models:
    pipeline=Pipeline([('PCA',PCA(n_components=4)),('model',model)])
    model.fit(X_train,y_train)
    print('---'*30,'\n',model,'\n','---'*30)
    #print(model.score(X_test,y_test))
    print(r2_score(model.predict(X_test),y_test))
    #print(classification_report(model.predict(X_test),y_test))
```

#### Other than KNN all Perform with agreeable accuracy but the number of principle components and consistancy through cross validation is yet to be explored.

```python
max_sc=0
optimal_dim=1
for i in range(1,9):
    pipe_lin=Pipeline([('PCA',PCA(n_components=i)),('linear',LinearRegression())])
    scl=cross_val_score(pipe_lin,x,y1)
    print(np.mean(scl),np.std(scl))
    sc=scl.mean()
    if sc>max_sc:
        max_sc=sc
        optimal_dim=i
print(max_sc,optimal_dim)
```

```python
max_sc=0
optimal_dim=1
for i in range(1,9):
    pipe_nn=Pipeline([('PCA',PCA(n_components=i)),('nn',MLPRegressor())])
    scl=cross_val_score(pipe_nn,x,y1)
    print(np.mean(scl),np.std(scl))
    sc=scl.mean()
    if sc>max_sc:
        max_sc=sc
        optimal_dim=i
print(max_sc,optimal_dim)
```

```python
max_sc=0
optimal_dim=1
for i in range(1,7):
    pipe_tree=Pipeline([('PCA',PCA(n_components=i)),('tree',DecisionTreeRegressor())])
    scl=cross_val_score(pipe_tree,x,y1)
    print(np.mean(scl),np.std(scl))
    sc=scl.mean()
    if sc>max_sc:
        max_sc=sc
        optimal_dim=i
print(max_sc,optimal_dim)
```

```python
max_sc=0
optimal_dim=1
for i in range(1,7):
    pipe_svr=Pipeline([('PCA',PCA(n_components=i)),('svr',SVR())])
    scl=cross_val_score(pipe_svr,x,y1)
    print(np.mean(scl),np.std(scl))
    sc=scl.mean()
    if sc>max_sc:
        max_sc=sc
        optimal_dim=i
print(max_sc,optimal_dim)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y1,test_size=0.20)
pipe_nn=Pipeline([('PCA',PCA(n_components=1)),('nn',MLPRegressor())])
pipe_nn.fit(X_train,y_train)
y_pred=pipe_nn.predict(X_test)
r2_score(y_pred,y_test)
```

#### After considering the mean cross validation score and the standard deviation among the scores we choose neural network with 1 principal component which has acceptable mean score and the score varies only by 10%.

```python
from sklearn.externals import joblib
joblib.dump(pipe_nn,'Covid_death.pkl')
```

```python
y1_pred=pd.DataFrame(y_pred)
y1_pred.to_csv('Covid_death.csv')
```
